# Exploratory Data Analysis

## Gather

```python
# Library Imports
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os
import sys

# Set Plot Style
sns.set_theme(style='whitegrid', palette='muted', font_scale=1.5)
%matplotlib inline

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data
heart_attack = pd.read_csv("../data/raw/heart_attack_dataset.csv")

# Display the shape of the dataset
print(f"Shape of the dataset: {heart_attack.shape}")

# Display the first 5 rows of the dataset
heart_attack.head()
```

## Assess

### Basic information about the dataset
- 1. Number of rows and columns
- 2. Column names
- 3. Data types
- 4. Missing values
- 5. Duplicates
- 6. Unique values
- 7. Summary statistics


```python
# Display the columns of the dataset
heart_attack.columns
```

```python
# Display the information of the dataset
heart_attack.info()
```

```python
# Display the summary statistics of the dataset
heart_attack.describe()
```

```python
# Check for missing values
heart_attack.isnull().sum()
```

```python
# Check for duplicates
heart_attack.duplicated().sum()
```

```python
# Summary statistics for non-numeric columns
heart_attack.describe(include='object')
```

```python
# Get the numerical columns and categorical columns of the dataset
numerical_cols = heart_attack.select_dtypes(include=[np.number]).columns
categorical_cols = heart_attack.select_dtypes(include=[object]).columns

print(f"Numerical Columns: {numerical_cols}")
print(f"Categorical Columns: {categorical_cols}")
```

### Findings
- 1. The dataset has 372974 rows and 32 columns.
- 2. The dataset contains a mix of numerical and categorical data.
- 3. There are no missing values in the dataset.
- 4. There are no duplicates in the dataset.
- 5. The dataset contains a mix of continuous and discrete data.


## Explore the dataset

```python
# Target Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="Outcome", data=heart_attack)
plt.xlabel("Heart Attack Outcome")
plt.ylabel("Count")
plt.title("Distribution of Heart Attack Outcome")
plt.savefig("../outputs/figures/target_distribution.png")
plt.show()
```
